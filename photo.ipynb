```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# CSV
data = pd.read_csv('m.csv')

# argmax()1
data['combined_group'] = np.argmax(data[['CS_Name_CS2_35', 'CS_Name_CS2_36', 'CS_Name_CS2_37', 'CS_Name_CS2_38']].values, axis=1)

# 
grouped_data = data.groupby('combined_group')

# 
for group_name, group_data in grouped_data:
    plt.plot(group_data['cycle'], group_data['SoH'], label=group_name)

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Your Title')
plt.legend()  # 
plt.show()
```

```python
for group_name, group_data in grouped_data:
    plt.plot(group_data['resistance'], group_data['SoH'], label=group_name)

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Your Title')
plt.legend()  # 
plt.show()
```

```python
for group_name, group_data in grouped_data:
    plt.plot(group_data['log_cycle'], group_data['SoH'], label=group_name)

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Your Title')
plt.legend()  # 
plt.show()
```
